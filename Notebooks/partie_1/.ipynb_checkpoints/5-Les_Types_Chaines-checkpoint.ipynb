## Manipulation des chaines

### Transformation en miniscules

```python
word = "HELLO World".lower()
word
```

### Transformation en majiscules

```python
word = "HELLO World".upper()
word
```

### Capitalisation  d'une phrase

```python
word = 'hello world'.capitalize()
word
```

### Title & Swap case

```python
word = 'hello world'.title()
print(word)
```

```python
word = 'Hello World'.swapcase()
print(word)
```

### Inverser une chaine

```python
word = 'Hello world'
rword = word[::-1]
rword
```

### Enlevement des espaces

```python
word = '      Hello world  '
print(word)
print(word.strip())
```

### Concaténaton des chaines

```python
word  = 'Hello ' + ' world'
print(word)
```

### Formattage des chaines

```python
nom = 'Bechir'
prenom ='Bejaoui'
age = 40

presentation = f'Mon nom est {nom}, mon prénom est {prenom} et mon age est {age}'
presentation
```

```python
nom = 'Bechir'
prenom ='Bejaoui'
age = 40
presentation = 'Mon nom est {0}, mon prénom est {1} et mon age est {2}'
presentation = presentation.format(nom,prenom,age)
presentation
```

```python
nom = 'Bechir'
prenom ='Bejaoui'
age = 40
presentation = 'Mon nom est {}, mon prénom est {} et mon age est {}'
presentation = presentation.format(nom,prenom,age)
presentation
```

```python
nom = 'Bechir'
prenom ='Bejaoui'
age = 40
presentation = 'Mon nom est %s, mon prénom est %s et mon age est %i'%('Bechir','Bejaoui',40)
presentation = presentation.format(nom,prenom,age)
presentation
```
