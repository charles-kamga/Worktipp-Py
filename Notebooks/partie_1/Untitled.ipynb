```python
def division(x,y):
    try:
        return x / y 
    except ZeroDivisionError as zde:
        print(zde.args)
```

```python
result = division(50,0)
```

```python
result = division(50,10)
```

```python
def division(x,y):
    try:
        res = x /  y
    except ZeroDivisionError as zde:
        print(zde.args)
    else:
        print(f'Le resultat est {res}')
```

```python
result = division(50,10)
```

### Le bloc finally

```python
def nom(moi):
    try:
        if moi != 'Béchir':
            raise NameError()
        else: 
            print('C\'est bien moi')
    except NameError as ne:
        print('Ce n\'est pas moi')
    finally:
        print('Nous arrivons à la fin du routine')
```

```python
nom('Béchir')
```

```python
nom('Hellen')
```

### La gestion d'execptions multiples

```python
try:
    a = -50
    b = 0
    c = a / b
    if not c > 0 & isinstance(c,int):
        raise ArithmeticError('c doit être un entier naturel')
        
    print(c)
except ArithmeticError as are:
    print(are.args)
except ZeroDivisionError as zde:
    print(zde.args)
except Exception as e:
    print(e.args)
```
