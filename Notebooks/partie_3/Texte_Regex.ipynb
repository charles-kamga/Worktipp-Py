### 2.2 Les recherches basées sur les assertions

Une assertion est un mode  de recherche qui cherche une correspondance si elle est trouvée ou non 

Les assertions sont basées sur: 
    1. Les Anchors
    2. Les Lookaround (Lookahead & Lookbehind)

#### 2.2.1 Les Anchors

```python
patterns = ['^\w+','\w+$']
for pattern in patterns:
    resultat = re.search(pattern,content)
    print(resultat)
```

#### 2.2.2 Les Lookaround

```python
texte ='Les expressions régulières sont du casse-tête'
```

__Lookahead__

```python
pattern = '(?<=-)\w+'
resultat = re.search(pattern,texte)
resultat
```

__Lookbehind__

```python
pattern = '\w+(?=-)'
resultat = re.search(pattern,texte)
resultat
```

### 2.3. Recherches basées sur le nombre de répetitions 

__Les quantificateurs ?,* et +:__


```python
texte = 'mama'
pattern = 'mama+'
match = re.search(pattern,texte)
if match:
    print('ça correspend')
else: 
    print('ça ne correspend pas')
```

__Quantité{N}__

```python
texte= '0123456789 est un entier long'
pattern ='\d{10}'
resultat = re.search(pattern,texte)
resultat
```

__Quantité{N,M}__

```python
texte= '0123456 est un entier long'
pattern ='\d{8,10}'
resultat = re.search(pattern,texte)
resultat
```

__Quantité{N,}__

```python
texte= '012 est un entier long'
pattern ='\d{8,}'
resultat = re.search(pattern,texte)
resultat
```

### 2.4. Recherches basées sur les groupes 

__Sans groupes__

```python
file = open('c:\\temp\\io\\emails.txt')
content = file.read()
content
```

```python
pattern = r'[A-Za-z]+@[A-Za-z]+.fr|[A-Za-z]+@[A-Za-z]+.com|[A-Za-z]+@[A-Za-z]+.ch'
resultat = re.findall(pattern,content)
print(resultat)
```

```python
pattern = r'(?<=@)[A-Za-z]+'
resultat = re.findall(pattern,content)
print(resultat)
```

```python
pattern = r'(?<=.fr)[A-Za-z]+(?=@)|(?<=.com)[A-Za-z]+(?=@)|(?<={.}ch)[A-Za-z]+(?=@)'
resultat = re.findall(pattern,content)
print(resultat)
```

```python
pattern = r'(fr|com|ch)'
resultat = re.findall(pattern,content)
print(resultat)
```

__Avec les groupes__

```python
pattern = '(([A-Za-z]+)@([A-Za-z]+).(fr|com|ch))'
resultat = re.findall(pattern,content)
print(resultat)
```

```python
resultat = re.findall(pattern,content)
```

```python
print(resultat)
```

```python
for element in resultat:
    print(element[2])
```

```python
iterateur = re.finditer(pattern,content)
```

```python
for element in iterateur:
    print(element.group(3))
```

### 3. La compilations

```python
regexes = [re.compile(p) for p in ['un','deux','trois']]
texte = 'J\'ai Un lapin, Deux poules et trois dindons'

for regex in regexes:
    if regex.search(texte):
        print(f'Le pattern {regex.pattern} est trouvé')
    else:
        print(f'Le pattern {regex.pattern} n\'est pas trouvé')
```
