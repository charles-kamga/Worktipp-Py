# Les dates

### Imprimer la date actuelle

```python
import datetime

date = datetime.datetime.now()
print(date)
```

### Créer une date

```python
date = datetime.datetime(2019,1,14,0,0,0)
print(date)
```

### Formatter la date

```python
date = datetime.datetime(2019,1,14,0,0,0)
print(date.strftime("%B"))
```

```python
import locale
locale.setlocale(locale.LC_TIME,'')
date = datetime.datetime(2019,1,14,0,0,0)
print(date.strftime("%B"))
```

# Time

```python
import time;

localtime = time.localtime(time.time())
print (f"Local current time : {localtime}")
```

```python
time.timezone
```

```python
struct_time = time.strptime("30 Nov 00", "%d %b %y")
print (f"returned tuple: %s {struct_time}" )
```

```python
print(f"time.altzone %d {time.altzone}")  
```
