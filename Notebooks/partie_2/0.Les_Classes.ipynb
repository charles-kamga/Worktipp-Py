### 1. Héritage

```python
class Homonoide(object):
    pass
```

```python
class Homonoide:
    def __init__(self,taille,age):
        self.taille = taille
        self.age = age
```

```python
class Humain(Homonoide):
    def __init__(self,nom,prénom,taille,age):
        super().__init__(taille,age)
        self.nom = nom
        self.prénom = prénom 
    def affiche(self):
        print(f'{self.nom} {self.prénom} {self.taille} {self.age}')
```

```python
moi = Humain('Béchir','Béjaoui',1.85,40)
moi.affiche()
```

### 1.1 Héritage Multiple

```python
class Personne:  
    def __init__(self, nom, age):  
        self.nom = nom  
        self.age = age  
        
    def PersonneInfo(self):  
        print(f'Nom: {self.nom} Age: {self.age} ')  
```

```python
class Employee:
    def __init__(self, nas):  
        self.nas = nas  
  
    def socialInfo(self):  
        return self.nas 
```

```python
class Resident(Personne, Employee): # extends both Person and Student class  
    def __init__(self, nom, age, nas):  
        Personne.__init__(self, nom, age)  
        Employee.__init__(self, nas)  
```

```python
resident = Resident('Bechir',40,'Permis de travail')
```

```python
resident.PersonneInfo()
resident.socialInfo()
```

### 2. Mangling (Encapsulation)

```python
class Personne:   
    def __init__(self, nom, age):  
        self.nom = nom  
        self.age = age  
        
    def PersonneInfo(self):  
        print(f'Nom: {self.nom} Age: {self.age} ')  
```

__Membres protégés par convention__

```python
class Personne:   
    def __init__(self, nom):  
        self.nom = nom  
        self._age = 40  
        
    def _personneInfo(self):  
        print(f'Nom: {self.nom} Age: {self._age} ')  
```

```python
moi = Personne('Bechir')
moi._age
moi._personneInfo()
```

__Membres privés par convention__

```python
class Personne:   
    def __init__(self, nom):  
        self.nom = nom  
        self.__age = 40  
        
    def __personneInfo(self):  
        print(f'Nom: {self.nom} Age: {self.__age} ')  
```

```python
moi = Personne('Bechir')
moi.__age
moi.__personneInfo()
```

## 3. Le plymorphisme 

```python
class Homonoide(object):
    def __init__(self,poids=0,taille=0,age=0):
        self.poids = poids
        self.taille = taille
        self.age = age
        
    def déplacer(self):
        
        return f'Un homonoide est entrain de se déplacer'
```

```python
class Humain(Homonoide):
    
    def __init__(self,nom,prénom):  
        self.nom = nom
        self.prénom = prénom
        super(Humain).__init__()
        
    def dépalcer(self):
        basem = super(Humain,self).déplacer()
        print(basem)
        return f'{self.nom} {self.prénom} est entrain de ce déplacer '
```

```python
Homonoide().déplacer()
```

```python
moi = Humain('Béchir','Béjaoui')
```

```python
moi.dépalcer()
```

### 4.2 Les méthodes statique vs méthodes de classe

```python
class Femme(Humain):
    
    Genre ='Féminin'
    
    def __init__(self,matrimoniale='Mademoiselle'):
        self.matrimoniale = matrimoniale
        print(f'C\'est une {matrimoniale}')
        
    def affiche(self,nom,prénom):
        print(f'{nom} {prénom}')

    @classmethod
    def afficheGenre(cls):
        print(cls.__name__)
        print(f'Le genre est {Femme.Genre}')
    
    @staticmethod
    def afficheGenre2():
        print(f'Le genre est {Femme.Genre}')
    
```

```python
Femme.afficheGenre2()
```
