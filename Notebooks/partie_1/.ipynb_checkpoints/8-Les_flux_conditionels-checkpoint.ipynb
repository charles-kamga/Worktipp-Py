## Les flux conditionels

### Le bloc if

a = 55
b = 200
if a < b : 
    print('a est supérieur à b')

### Le bloc if else

```python
a = 55
b = 200
if a < b : 
    print('a est inférieur à b')
else:
    print('a est supérieur à b')
```

```python
a = 1155
b = 200
if a < b : 
    print('a est inférieur à b')
else:
    print('a est supérieur à b')
```

### Une autre alternative au bloc if else

```python
a = 55
b = 100
result = ('a est inférieur à b','a est supérieur à b')[a>b]
print(result)
```

On appelle souvent cette expression __terany conditional operator__

### Le bloc if elif else

```python
couleur = 'vert'

if couleur =='rouge':
    print('La couleur est rouge')
elif couleur =='bleu':
    print('La couleur est bleu')
elif couleur =='vert':
    print('La couleur est vert')
else: 
    print('La couleur ne figure pas dans la liste des couleurs primitives')
```
