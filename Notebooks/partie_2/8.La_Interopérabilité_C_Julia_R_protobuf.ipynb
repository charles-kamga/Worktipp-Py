## 5. Protobuff

1. __Buffer protocol__ est un __Framework__ qui permet à __Python__ de traiter des volumes et des structures de données représentées sous forme de tableaux d'octets 

2. __Buffer protocol__ est très utile pour les datasciences du faite qu'il est possible que plusieurs codes Python peuvent traiter le même tampons de données souvent volumineu sans avoir recour à faire des copies

3. __Buffer protocol__ permet aussi l'interopérabilité avec des languages bas niveau essentiellement __C/C++__

4. __Buffer protocol__ est souvent connu par l'abréviation [PEP 3118](https://legacy.python.org/dev/peps/pep-3118/) 

Protocol Buffers est un format de sérialisation avec un langage de description d'interface développé par Google dont le but principal est l'interopérabilité de python avec d'autres languages tel que __C++__,__C#__ , __Objective-C__ , __Java__,__Ruby__ et __GO__

__Avantages__:<br>
    1. Presque 6 fois moins volumineux que XML
    2. Presque 60 fois plus rapide à utiliser
    3. Convenable pour les systèmes embarquées & les capteurs
