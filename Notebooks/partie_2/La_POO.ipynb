## 1. Héritage

```python
class Homonoide(object):
    pass
```

```python
class Homonoide:
    def __init__(self,poids,taille,age):
        self.poids = poids
        self.taille = taille
        self.age = age
        
    def affiche(self):
        print(f'{self.poids} ; {self.taille}; {self.age}')
```

```python
class Humain(Homonoide):
    def __init__(self,nom,prénom,poids,taille,age):
        self.nom = nom
        self.prénom = prénom 
        Homonoide.__init__(self,poids,taille,age)
```

```python
moi = Humain('Bechir','Bejaoui',96,1.85,40)
```

```python
moi.affiche()
```

### 1.1. Héritage multiple

```python
class Personne:
    def __init__(self,nom,age):
        self.nom = nom
        self.age = age
    def personneInfo(self):
        print(f'{self.nom}  {self.age}')
```

```python
class Employee:
    def __init__(self,nas):
        self.nas = nas
    def employeeInfo(self):
        print(f'{self.nas}')
```

```python
class Resident(Personne,Employee):
    def __init__(self,nom,age,nas):
        Personne.__init__(self,nom,age) 
        Employee.__init__(self,nas)
```

```python
resident001 = Resident('Bechir',40,123456789)
```

```python
resident001.personneInfo()
resident001.employeeInfo()
```

__Le membre protégé sous Python__

```python
class Personne : 
    def __init__(self,nom,_age=40):
        self.nom = nom 
        self._age = _age
   
    def tostring(self):
        print (f'{self.nom} {self._age}')
```

```python
moi = Personne('Bechir',40)
moi.tostring()
```

## 3. Polymorphisme

```python
class Homonoide:
    def __init__(self):
        print('Je suis un Homonoide')
    def se_déplacer(self):
        print('Un homonoide ce déplace')
```

```python
class Humain(Homonoide):
    def __init__(self,nom):
        self.nom = nom
        print(f'Je suis {self.nom}')
    def se_déplacer(self):
        super().se_déplacer()
        print(f'{self.nom} se déplace')
```

```python
moi = Humain('Béchir Béjaoui')
moi.se_déplacer()
```

```python
class Femme(Humain):
    Genre = 'Féminin'
    def __init__(self,nom,prénom, matrimoniale='Mademaosielle'):
        Humain.__init__(self,nom,prénom)
        self.matrimoniale = matrimoniale
        
    def affiche(self):
        print(f'{self.matrimoniale} {self.nom}  {self.prénom} ')
    
    @classmethod
    def afficheGenre(cls):
        print(cls.__name__)
        print(f'Le genre est {Femme.Genre}')
```

```python
class Femme(Humain):
    Genre = 'Féminin'
    def __init__(self,nom,prénom, matrimoniale='Mademaosielle'):
        Humain.__init__(self,nom,prénom)
        self.matrimoniale = matrimoniale
        
    def affiche(self):
        print(f'{self.matrimoniale} {self.nom}  {self.prénom} ')
    
    @classmethod
    def afficheGenre(cls):
        print(cls.__name__)
        print(f'Le genre est {Femme.Genre}')
    
    @staticmethod
    def afficheGenrestatique():
        print(f'Le genre est {Femme.Genre}')
```
