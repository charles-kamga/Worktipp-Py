```python
from random import randint as r

randomList = [r(0,1000),r(1001,2000),r(2001,3000),r(3001,5001),r(6001,10000)]
print(randomList)
data = []
l= []
for i in range(1,10000):
    l = [r(1,5),r(10,50),r(3,20),r(5,55),r(10,65)]
    data.append(l)
    if i in randomList:
        data.append(l)
        randomList.remove(i)
print(randomList) 
columns  = ['champ1','champ2','champ3','champ4','champ5']
df = p.DataFrame(data,columns=columns)
df
```
