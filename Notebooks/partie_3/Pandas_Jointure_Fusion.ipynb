```python
import pandas as p

data1 = {
    'id' : [1,2,3,4,5],
    'data':['abc','bcd','cdh','dhi','hij']
}

data2 = {
    'id' : [6,4,3,1,7],
    'data':['ijk','jkl','klm','lmn','mno']
}

data3 = {
    'id' : [8,9,10,11,12],
    'data':['nop','opq','pqr','qrs','rst']
}

df1 = p.DataFrame(data1)
df2 = p.DataFrame(data2)
df3 = p.DataFrame(data3)
```

### Concatenation

```python
df_concat = p.concat([df1,df2],axis=0)
df_concat
```

```python
df_concat = p.concat([df1,df2],axis=1)
df_concat
```

### Fusion ou Jointure

```python
df_merge = p.merge(df1,df2,on='id')
df_merge
```

```python
df_merge = p.merge(df1,df2,left_on='id',right_on='id')
df_merge
```

```python
df_merge_outer = p.merge(df1,df2,on='id',how='outer')
df_merge_outer
```

```python
df_merge_outer_left = p.merge(df1,df2,on='id',how='left')
df_merge_outer_left
```

```python
df_merge_outer_rigth = p.merge(df1,df2,on='id',how='right')
df_merge_outer_rigth
```

```python
df12 = p.merge(df1,df2,on='id', how='outer')
```

```python
df123 = p.merge(df12,df3,on='id',how='outer')
```
