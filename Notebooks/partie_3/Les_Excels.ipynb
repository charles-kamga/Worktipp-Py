## 1.Le module openpyxl

#### 1.1 Obtenir la valeur d'une cellule Excel

```python
from openpyxl import load_workbook

path ='C:\\temp\\io\\FL_insurance_sample.xlsx'

wb = load_workbook(path)

feuille = wb.active
cellule  = feuille.cell(row=1,column=1)

print(cellule.value)
```

#### 1.2 Obtenir la valeur d'une plage cellules

```python
from openpyxl import load_workbook

path ='C:\\temp\\io\\FL_insurance_sample.xlsx'

wb = load_workbook(path)

feuille = wb.active

header1 = feuille.cell(row = 1,column = 1)
header2 = feuille.cell(row = 1,column = 2)
header3 = feuille.cell(row = 1,column = 3)
print(f'{header1.value}  {header2.value}  {header3.value}')

for i in range(2,10):
    cell1 = feuille.cell(row = i,column = 1)
    cell2 = feuille.cell(row = i,column = 2)
    cell3 = feuille.cell(row = i,column = 3)
    print(f'{cell1.value}  {cell2.value}  {cell3.value}')
```

#### 1.3 Obtenir le nombre total des lignes et de colonnes

```python
from openpyxl import load_workbook

path ='C:\\temp\\io\\FL_insurance_sample.xlsx'

wb = load_workbook(path)

feuille = wb.active

nb_colonnes = feuille.max_column
nb_lignes = feuille.max_row

print(f'Le nombre maximal des colonnes {nb_colonnes}')
print(f'Le nombre maximal des lignes {nb_lignes}')
```

## 2.Le module xlrd

#### 1.1 Obtenir la valeur d'une cellule Excel

```python
from openpyxl import load_workbook
path = 'C:\\temp\\io\FL_insurance_sample.xlsx'

wb = load_workbook(path)
feuille = wb.active
nb_colonnes = feuille.max_column
nb_lignes = feuille.max_row

print(f'Le nombre maximal des colonnes est {nb_colonnes}')
print(f'Le nombre maximal des lignes est {nb_lignes}')
```

#### 1.2 Obtenir la valeur d'une plage cellules

```python
from xlrd import open_workbook

path = 'C:\\temp\\io\FL_insurance_sample.xlsx'
wb = open_workbook(path)
feuille = wb.sheet_by_index(0)

for r in range(0,10):
        cell1 = feuille.cell_value(r,0)
        cell2 = feuille.cell_value(r,1)
        cell3 = feuille.cell_value(r,2)
        print(f'{cell1} {cell2} {cell3}\n')
```

#### 1.3 Obtenir le nombre total des lignes et de colonnes

```python
from xlrd import open_workbook

path ='C:\\temp\\io\\FL_insurance_sample.xlsx'

wb = open_workbook(path)
feuille = wb.sheet_by_name('FL_insurance_sample')

nb_colonnes= feuille.ncols 
nb_lignes = feuille.nrows

print(f'Le nombre total des colonnes est {nb_colonnes}')
print(f'Le nombre total des lignes est {nb_lignes}')
```
