```python
path = 'C:\\temp\\io\\classe.txt'
kwargs = dict(missing_values="N/A", names=False, case_sensitive=True)
data = np.recfromcsv(path,delimiter=',')
print(data)
```

```python
from matplotlib import pyplot as plt

moyennes = []

for notes in data:
    moyenne = 0
    notes = [float(i) for i in notes]
    moyenne = np.mean(notes)
    moyennes.append(moyenne)

moyennes
```

```python
plt.xlabel('eleves')
plt.ylabel('Moyennes')
plt.xticks(range(0,20))
plt.plot(moyennes)
```

```python
liste = np.random.randn(999999)
```

```python
def calcul(l):
    longeur = len(l)
    l2 = np.zeros(longeur-1)
    for i in range(len(l2)):
        l2[i]= 2*l[i+1]-3*l[i]
    return l2
```
