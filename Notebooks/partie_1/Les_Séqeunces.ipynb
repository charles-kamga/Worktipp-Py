##### Les dictionnaires:
formes d'une paire (cle,valeur). Les valeurs peuvent être de n'importe quel type de valeur. Une fois qu'une clef est ajoutee au dico, elle est immuable.

```python
dict1 = {'1':'un','2':'deux','3':'tois'}
```

```python
dict2 = {'1':'un','2': 2,'3':True}
```

```python
dict3 = {1 :'un','2': 2,'3':True}
```

```python
dict3
```

```python
# Lister les clefs d'un dictionnaire
dict3.keys()
```

```python
# Lister les valeurs d'un dictionnaire
dict3.values()
```

```python
# Lister les paires (cles,valeurs)
dict3.items()
```

```python
dict1
```

```python
# Mettre a jour un dictionnaire en changeant la valeur d'une clef
dict1.update({'1' : 'one'})
```

```python
dict1
```

```python
# Mettre a jour un dictionnaire en ajoutant une paire (cle,valeur)
dict1.update({'uno' : 'one'})
```

```python
dict1
```

```python
# trier un dico par ses clefs.
sorted(dict1.items())
```
