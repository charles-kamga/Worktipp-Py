```python
import pandas as p 

data = [[1,2,3,51,8],[5,4,3,6,7],[4,2,8,1,5],[1,2,5,8,7]]

cols = ['champ1','champ2','champ3','champ4','champ5']

df = p.DataFrame(data,columns=cols)
```

```python
df
```

```python
df.plot(kind='scatter', x=['champ1','champ2'], y=['champ3','champ4'])
```

```python
import numpy as np
seuil = (np.percentile(df,75)-np.percentile(df,25))+ df.champ4.mean()
```

```python
seuil
```

```python
df['champ4'] = np.where(df.champ4>=seuil,int(df.champ4.mean()),df.champ4)
```

```python
df
```

```python
seuil = 1.5*(np.percentile(df,75)-np.percentile(df,25))
```

```python
seuil
```

```python
df.drop(df[df['champ4'] > seuil + np.percentile(df['champ4'], 75)].index
    ,inplace=True)
df.drop(df[df['champ4'] < np.percentile(df['champ4'], 25)-seuil].index
    ,inplace=True)
```
