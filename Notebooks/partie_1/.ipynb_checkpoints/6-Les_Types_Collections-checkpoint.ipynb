### Les dictionaires

Représentent une forme spéciale de collection, ou les éléments sont composé d'une paire clé/valeur

&emsp;&emsp;&emsp;1. Les clés doivent être uniques et immutables dans un dictionnaire<br>
&emsp;&emsp;&emsp;2. Les valeurs peuvent être de n'importe quel type<br>
&emsp;&emsp;&emsp;3. Contrairement aux listes qui utilisent [ ]et les tuples qui utilisent ( ), les dictionaires utilisent { }<br>

#### Accéder à une valeur dans le dictionnaire

```python
dic1 = {'1':'un','2':'deux','3':'trois'}
dic1['1']
```

```python
dic1 = {1:'un',2:'deux',3:'trois'}
dic1[1]
```

#### Enumération des clés

```python
dic1.keys()
```

#### Enumération des valeurs

```python
dic1.values()
```

#### Enumération des éléments

```python
dic1.items()
```

#### Suprimmer un élément sur la base de clé 

```python
del dic1[1]
```

```python
dic1.items()
```

#### Mettre à jour ou ajouter un élément à un dicionaire

```python
dic1.update({1:'un'})
```

```python
dic1.items()
```

#### Trier les élements du dictionnaire

```python
sorted(dic1.items())
```

```python
help({ })
```

La différence entre A et B (A - B) est un ensemble d’éléments qui ne figurent que dans A mais pas dans B. De même, B - A est un ensemble d’éléments dans B mais pas dans A.

```python
from IPython.display import Image
from IPython.core.display import HTML 
Image(url= "http://localhost:8888/notebooks/Documents/Python_Scripts/figures/set-difference.jpg")
```

```python
A = {1, 2, 3, 4, 5}
B = {4, 5, 6, 7, 8}
print(A - B)
```
