### Le passage d'un nombre indeterminé de paramètres 

```python
def affiche(* args):
    for element in args:
        print(element)
    
```

```python
liste = [1,[1,2,3,],False,'Hello world']
affiche(liste)
```

```python
def affiche(* args, x):
    for element in args:
        print(element)
```

```python
affiche (liste,'Béchir')
```

```python
def affiche(x, * args):
    print(x)
    for element in args:
        print(element)
```

```python
affiche ('Béchir', liste)
```

```python
def affiche(x,*args,**kwargs):
    print(x)
    for info in kwargs:
        print(info)
    
```

```python
def factoriel(n):
    #Effectuer des vérifications nécessaires pour que n soit un entier naturel
    if not isinstance(n,int):
        raise TypeError('n doit être un numérique et plusp précisement un entier naturel')
    if not n >= 0:
        raise TypeError('n ne doit pas être négatif')
   
    def factoriel_cuisine_interne(n):
        if n<=1:
            return 1 
        else:
            return n * factoriel_cuisine_interne(n-1)
    return factoriel_cuisine_interne(n)
```

```python
resultat = factoriel(6)
```

```python
print(resultat)
```
