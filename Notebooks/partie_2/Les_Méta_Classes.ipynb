### 1. Usage pour méta classe personalisée

```python
class Personne_Meta(type):
     def __new__(cls,nom,bases,membres):
        print(f'Entrer dans la portée de __new__')
        if nom == 'NoneType':
            return None 
        print(cls)
        print(nom)
        print(bases)
        for membre in membres:
            print(membre)
        return super(Personne_Meta,cls).__new__(cls,nom,bases,membres)
    
   
     def __init__(self,nom,bases,membres):
        print(f'Prè appel d\'init')
        super(Personne_Meta,self).__init__(nom,bases,membres)
    
     def __call__(cls,*args,**kwargs):
        print(f'Appel de __call__')
        instance = super(Personne_Meta,cls).__call__(*args,**kwargs)
        print(instance.nom)
        return instance
```

```python
class Homonoide: 
    pass
```

```python
class Personne(Homonoide,metaclass=Personne_Meta):
    def __init__(self,nom,age,poids,taille):
        self.__nom = nom 
        self.__age = age 
        self.__poids = poids 
        self.__taille = taille 
    
    @property
    def nom(self):
        return self.__nom
    
    @nom.setter
    def nom(self,valeur):
        self.__nom = valeur
        
    @property
    def age(self):
        return self.__age
    
    @age.setter
    def age(self,valeur):
        self.__age = valeur
    
    @property
    def poids(self):
        return self.__poids
    
    @poids.setter
    def poids(self,valeur):
        self.__poids = valeur
        
    @property
    def taille(self):
        return self.__taille
    
    @taille.setter
    def taille(self,valeur):
        self.__taille = valeur
    
```

```python
moi = Personne('Béchir',40,90,1.85)
```
