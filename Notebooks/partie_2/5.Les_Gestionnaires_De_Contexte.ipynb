## Creation de gestionaire de contextes

Les __gestionnaires de contexte__  permettent d'allouer et de libérer d'une manière précise. L'exemple le plus largement utilisé de sont les gestionnaires de contexte  __with__

Les gestionaires de contexte sont utilisés principalement lors d'invocation des ressources externes au programmes tel que __les connexions au bases de données__ , __les opérations entrées / sorties__ et __la gestion des threads et process__

Garder ces mot clé ou les gestionnaires de contextes sont très utiles<br>
        1.Open - Close<br>
        2.Lock - Release<br>
        3.Change - Reset<br>
        4.Enter - Exit<br>
        5.Start - Stop

__Une première approche__

```python
class Ressource:
    def __init__(self):
        self.etat = 'Etat: Initialisation de ressource'
        print(self.etat)
              
    def ouvrir(self):
        self.etat = 'Etat: Début utilisation de ressource'
        
    def fermer(self):
        self.etat = 'Etat: Fin d\'utilisation de ressource'
    
    def __enter__(self):
        self.etat = 'Etat: Utilisation de ressource'
        print(self.etat)
        return self
    
    def __exit__(self,extype, valeur, extrace):
        self.etat = 'Etat: Liberation des ressources supposée faite maintenant'
        print(self.etat)
```

```python
ressource = Ressource
def use():
    try:
        res  = ressource()
    except: 
        print('Une exception est levées')
        res = None
    finally:
        res = None
        print('Liberation des ressources au niveau du bloc finally')
```

```python
use()
```

```python
def use():
    try:
        res  = ressource()
        raise Exception('Une execption est levée')
    except Exception as ex: 
        print(ex.__dict__)
        res = None
    finally:
        res = None
        print('Liberation des ressources')
```

```python
use()
```

__Une deuxième approche__ (Qui utilise le gestionnaire du contexte cette fois ci)

```python
with ressource() as res:
    print('Uilisation de la ressource')
```
