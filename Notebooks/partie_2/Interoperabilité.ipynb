```python
%%writefile c:\temp\Factoriel\FactorielModule.c
#include <Python.h>

int CFactoriel(int n)
{
   if(n<0)
    return -1
    else{
          if(n==0 || n == 1) return 1
          else
          {
              return n*CFactoriel(n-1);
          }
        }
}

/*Envolppes nécessaires*/
static PyObject* factoriel(PyObject* self,PyObject* args)
{
    int n;
    if(!PyArg_ParseTuple(args,"i",&n)) return NULL;
    return Py_BuildValue("i",CFactoriel(n));
}

static PyObject* version(PyObject* self)
{
    return Py_BuildValue("s","Version 1.0");
}


/*Structure représentatrice des wrappers*/

static PyMethodDef methods [] =
{
    {"factoriel",factoriel,METH_VARARGS,"Calcul de factoriel"},
    {"version",version,METH_NOARGS,"Calcul de factoriel"},
    {NULL,NULL,0,NULL}
};

PyModuleDef factorielModule = {PyModuleDef_HEAD_INIT,
                               "factorielModule",
                               "Facotiel Module",
                               -1,
                               methods};

PyMODINIT_FUNC PyInit_factorielModule(void)
{
    return PyModule_Create(&factorielModule);
}
```

```python
from distutils.core import setup,Extension
```

```python
%%writefile c:\temp\Factoriel\setup.py
from distutils.core import setup,Extension
module = Extension("factorielModule",sources=["FactorielModule.c"])
setup(name="FactorielPackage",version="1.0",description="Calcul de factoriel", ext_modules=[module])
```
