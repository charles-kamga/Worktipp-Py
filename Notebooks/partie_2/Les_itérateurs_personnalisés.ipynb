```python
class Personne: 
    def __init__(self,nom,prénom):
        self.nom = nom 
        self.prébom = prénom 
    
    def __str__(self):
        return self.nom 
```

```python
moi = Personne('Béchir','Béjaoui')
mafemme = Personne('Hellen','Béjaoui')
monfils = Personne('Bédis','Béjaoui')
mafille =Personne('Dihia','Béjaoui')

mafamille = Famille([moi,mafemme,monfils,mafille])
```

```python
class Famille: 
    __i = 0
    def __init__(self,membres):
        self.membres = membres 
    
    def __iter__(self):
        return self
    
    def __getitem(self,i):
        return self.membres[i]
    
    def __next__(self):
        if self.__i > len(self.membres)-1:
            raise StopIteration
        else: 
            self.__i +=1
            return self.membres[self.__i - 1]
```

```python
moi = Personne('Béchir','Béjaoui')
mafemme = Personne('Hellen','Béjaoui')
monfils = Personne('Bédis','Béjaoui')
mafille =Personne('Dihia','Béjaoui')

mafamille = Famille([moi,mafemme,monfils,mafille])
```

```python
for membre in mafamille:
    print(str(membre))
```

## 2. Générateurs

```python
class Famille: 
   
    def __init__(self,membres):
        self.membres = membres 
    
    def generateur(self):
        for membre in self.membres:
            yield membre
        
```

```python
moi = Personne('Béchir','Béjaoui')
mafemme = Personne('Hellen','Béjaoui')
monfils = Personne('Bédis','Béjaoui')
mafille =Personne('Dihia','Béjaoui')

mafamille = Famille([moi,mafemme,monfils,mafille])
```

```python
liste = mafamille.generateur()
```

```python
for element in liste: 
    print(element)
```

__Les générateurs sous forme d'expressions lambda__

```python
liste = [1,2,3,4,5,6,7,8,9]
```

```python
expression = lambda: [(yield element) for element in liste]
```

```python
print( *expression())
```
